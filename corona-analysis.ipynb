```python
import pandas as pd
import requests
```

```python
def read_worldometer_web_page():
    url = "https://www.worldometers.info/coronavirus/"
    response_obj = requests.get(url);
    worldometer_text = response_obj.text;
    return worldometer_text
```

```python
worldometer_text = read_worldometer_web_page()
```

```python
# uncomment below line to see extracted html data
# worldometer_text
```

```python
data = pd.read_html(worldometer_text)
```
